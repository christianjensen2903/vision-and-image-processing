```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_gradient_magnitude

image = plt.imread('peppers.jpg').mean(axis=-1)

sigmas = [1, 2, 4, 8]

fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.flatten()

for i, sigma in enumerate(sigmas):
    grad_magnitude = gaussian_gradient_magnitude(image, sigma=sigma, truncate=3.8)
    axes[i].imshow(grad_magnitude, cmap='gray')
    axes[i].set_title(f'Gradient Magnitude\n = {sigma}')
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
