## Vector Matrix Calculations, pen and paper

```python
from sympy import symbols, Matrix, simplify

b, a = symbols('b a')
v = Matrix([-2, b, 7, 2, a])

v_T_v = v.dot(v)

# Compute v * v^T
v_v_T = v * v.T

print(f"v^T * v = {v_T_v}")
print(f"v * v^T = {v_v_T}")
```

```python
x, y = symbols('x y')

B = Matrix([[-4, 8], [-5, -9], [4, -7]])

A = Matrix([[-5, 4, -6], [3, 4, -6], [8, -6, 8]])

v = B * Matrix([x, y])

print(f"v = {v}")

w = A * v

print(f"w = {w}")

# C = A.B
C = A * B

print(f"C = {C}")

z = C * Matrix([x, y])

print(f"z = {z}")
```

```python
x = symbols('x')

A = Matrix([[1, -1, -1], [8 - x, x - 6, x- 6], [x-7, 5-x, 5-x]])

# Compute A^2 and A^3
A_2 = (A * A).applyfunc(simplify)
A_3 = (A_2 * A).applyfunc(simplify)

print(f"A^2 = {A_2}")
print(f"A^3 = {A_3}")
```

```python
import numpy as np

A = np.array([
    [-95, 72, -4],
    [-144, 109, -6],
    [76, -57, 3]
])

# Define the vectors e1, e2, e3
e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])

# Solve the linear systems to find f1, f2, f3
f1 = np.linalg.solve(A, e1)
f2 = np.linalg.solve(A, e2)
f3 = np.linalg.solve(A, e3)

# Construct the matrix B
B = np.column_stack([f1, f2, f3])

# Compute A.B and B.A
AB = np.dot(A, B)
BA = np.dot(B, A)

# Print the results
print(f"B: {B}")
print(f"A * B: {AB}")
print(f"B * A: {BA}")
```

### Explanation of the Results:
- \( A \cdot B \): This should return the identity matrix since \( \mathbf{f}_1, \mathbf{f}_2, \mathbf{f}_3 \) were calculated to satisfy \( A \mathbf{f}_i = \mathbf{e}_i \).
- \( B \cdot A \): This operation will reveal if \( B \) is an inverse of \( A \) (it should also return the identity matrix if \( A \) is invertible).

To show that \( A^3 \) is always skew-symmetric, we rely on the definition of a skew-symmetric matrix and the properties of matrix multiplication and transposition.

### 1. Definition of a Skew-Symmetric Matrix:
A matrix \( A \) is skew-symmetric if:
\[
A^T = -A
\]

### 2. Consider \( A^3 = A \cdot A \cdot A \):
We aim to show that \( (A^3)^T = -A^3 \), which would confirm that \( A^3 \) is skew-symmetric.

### 3. Transpose of \( A^3 \):
Using the property of transposes \( (BC)^T = C^T B^T \), we compute:
\[
(A^3)^T = (A \cdot A \cdot A)^T = A^T \cdot A^T \cdot A^T
\]

### 4. Replace \( A^T \) with \( -A \):
Since \( A \) is skew-symmetric, \( A^T = -A \). Substituting this:
\[
(A^3)^T = (-A) \cdot (-A) \cdot (-A)
\]

### 5. Simplify:
\[
(A^3)^T = -A \cdot A \cdot A = -(A \cdot A \cdot A) = -A^3
\]

## Derivatives

```python
import matplotlib.pyplot as plt

def f(x):
    return np.exp(-x**2 / 2)

def f_prime(x):
    return -x * np.exp(-x**2 / 2)

def f_double_prime(x):
    return (x**2 - 1) * np.exp(-x**2 / 2)

# Plot f(x), f'(x), and f''(x) on the same graph
x = np.linspace(-7, 7, 500)
y_f = f(x)
y_f_prime = f_prime(x)
y_f_double_prime = f_double_prime(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y_f, label="$f(x) = e^{-x^2 / 2}$")
plt.plot(x, y_f_prime, label="$f'(x)$")
plt.plot(x, y_f_double_prime, label="$f''(x)$")
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.title("Function $f(x)$ and its derivatives")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x, t):
    return np.exp(-x**2 / (2 * t)) / np.sqrt(2 * np.pi * t)

# Define the functions
def df_dx(x, t):
    return -x / (t * np.sqrt(2 * np.pi * t)) * np.exp(-x**2 / (2 * t))

def df_dt(x, t):
    term1 = (x**2 / (2 * t**2 * np.sqrt(2 * np.pi * t))) - (1 / (2 * np.sqrt(2 * np.pi)**3))
    return term1 * np.exp(-x**2 / (2 * t))

def d2f_dx2(x, t):
    term1 = np.exp(-x**2 / (2 * t))
    term2 = (1 - x**2 / t)
    return -(1 / (t * np.sqrt(2 * np.pi * t))) * term1 * term2

# Values of t
t_values = [1, 2, 4]
x_values = np.linspace(-15, 15, 500)

plt.figure(figsize=(18, 12))
for i, t in enumerate(t_values):
    plt.subplot(3, 1, i + 1)
    plt.plot(x_values, f(x_values, t), label="$f(x,t)$")
    plt.plot(x_values, df_dx(x_values, t), label="$\\frac{\\delta f}{\\delta x}(x,t)$")
    plt.plot(x_values, df_dt(x_values, t), label="$\\frac{\\delta f}{\\delta t}(x,t)$")
    plt.plot(x_values, d2f_dx2(x_values, t), label="$\\frac{\\delta^2 f}{\\delta x^2}(x,t)$")
    plt.title(f"Functions for t={t}")
    plt.xlabel("x")
    plt.ylabel("Function Value")
    plt.legend()
    plt.grid()

plt.tight_layout()
plt.show()
   
```
